```python
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import time
from IPython import display
# %matplotlib notebook
```

```python
def test():
    ys = []
#     plt.ion()
    fig = plt.figure(1);
    for i in tqdm(range(100)):
        y = i ** 2 +2
        ys.append(y)
#         plt.clf()
        plt.plot(ys)
        display.clear_output(wait=True)
        plt.pause(0.1)
        plt.show()
#     plt.ioff()
```

```python

test()

# plt.show()
```

```python
import matplotlib.pyplot as plt

x = list(range(1, 100))  # epoch array
loss = [10 / (i**2) for i in x]  # loss values array

plt.ion()

for i in range(1, len(x)):
    ix = x[:i]
    iy = loss[:i]
    plt.title("loss")
    plt.plot(ix, iy)

    plt.xlabel("epoch")
    plt.ylabel("loss")

    # plt.xlim(0,len(x)) #x
    if i == 1:
        plt.pause(1)  # 
    plt.pause(0.5)

plt.ioff()
plt.show()
```
